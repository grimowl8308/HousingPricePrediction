## 3.2 Data Cleaning: handling missing values

### Remove features with over 25% missing values

```python
nan_mean = combined_data.isna().mean() * 100
nan      = nan_mean[nan_mean > 25].sort_values(ascending = False)

print("Columns with missing values above 25%:")
print(nan)
```

```python
def remove_excess_missing(data, upper_limit = 25):
    #purpose: remove features with too many missing values

    nan_mean = data.isna().mean()*100
    nan = nan_mean[nan_mean > upper_limit].sort_values(ascending = False)

    return data.drop(columns = nan.index)

training_data = remove_excess_missing(combined_data,25)
combined_data = training_data
```

### Missing Values, Unique Values, Dtype, Mean and Median
We found out that among all numerical features, **LotFrontage, GarageYrBlt has normal distribution**, so it is our intention to fill the missing values with their means

```python
# Display features with missing values
missing_values = combined_data.isnull().sum()
missing_values = missing_values[missing_values > 0]

# Create a DataFrame to store information about missing values
missing_info = pd.DataFrame(index=missing_values.index)
missing_info['Missing Val'] = missing_values
missing_info['Unique Val'] = combined_data[missing_values.index].nunique()
missing_info['DType'] = combined_data[missing_values.index].dtypes

# Filter features with numeric data types
numeric_features = combined_data[missing_values.index].select_dtypes(include=['float64', 'int64']).columns

# Calculate mean, median, and mode only for numeric features
missing_info.loc[numeric_features, 'Mean'] = combined_data[numeric_features].mean().round(2)
missing_info.loc[numeric_features, 'Median'] = combined_data[numeric_features].median().round(2)
# missing_info.loc[numeric_features, 'Mode'] = combined_data[numeric_features].mode().iloc[0].round(2)

# Sort the output by 'DType'
missing_info = missing_info.sort_values(by='DType')

# Display the information about missing values for numeric features
print("Features with Missing Values:")
print(f"\n{missing_info}")
```

### Replace missing values for LotFrontage, GarageYrBlt and GarageArea with their means

```python
combined_data['LotFrontage'].fillna(combined_data['LotFrontage'].mean(), inplace=True)
combined_data['GarageYrBlt'].fillna(combined_data['GarageYrBlt'].mean(), inplace=True)
```

### Recheck Missing Values, Unique Values, Dtype, Mean and Median

```python
# Display features with missing values
missing_values = combined_data.isnull().sum()
missing_values = missing_values[missing_values > 0]

# Create a DataFrame to store information about missing values
missing_info = pd.DataFrame(index=missing_values.index)
missing_info['Missing Val'] = missing_values
missing_info['Unique Val'] = combined_data[missing_values.index].nunique()
missing_info['DType'] = combined_data[missing_values.index].dtypes

# Filter features with numeric data types
numeric_features = combined_data[missing_values.index].select_dtypes(include=['float64', 'int64']).columns

# Calculate mean, median, and mode only for numeric features
missing_info.loc[numeric_features, 'Mean'] = combined_data[numeric_features].mean().round(2)
missing_info.loc[numeric_features, 'Median'] = combined_data[numeric_features].median().round(2)
# missing_info.loc[numeric_features, 'Mode'] = combined_data[numeric_features].mode().iloc[0].round(2)

# Sort the output by 'DType'
missing_info = missing_info.sort_values(by='Missing Val')

# Display the information about missing values for numeric features
print("Features with Missing Values and Additional Information:")
print(f"\n{missing_info}")
```

### Handling nulls for categorical / non-numerical features

```python
columns_to_analyze = ['GarageType','GarageFinish','GarageQual','GarageCond']

results = []

for column in columns_to_analyze:
    # Group by values of the column and calculate mean and median of SalePrice for each value
    grouped_data = combined_data.groupby(column)["SalePrice"].agg(["mean", "median"])
    
    for value, data in grouped_data.iterrows():
        results.append([column, value, round(data['mean'], 2), round(data['median'], 2)])

results_df = pd.DataFrame(results, columns=["Column", "Value", "Mean", "Median"])

print(results_df)
```

```python
columns_to_analyze = ['GarageType','GarageFinish','GarageQual','GarageCond']

for column in columns_to_analyze:
    # Find rows with null values in the current column
    null_rows = combined_data[combined_data[column].isnull()]
    
    # Group by values of the current column and calculate mean of SalePrice for each value
    mean_sale_price_by_value = combined_data.groupby(column)["SalePrice"].mean()
    
    # Iterate over each null row
    for index, row in null_rows.iterrows():
        # Find the nearest mean SalePrice from the same column
        nearest_mean = mean_sale_price_by_value.loc[mean_sale_price_by_value.sub(row["SalePrice"]).abs().idxmin()]
        
        # Find the value corresponding to the nearest mean SalePrice
        nearest_value = mean_sale_price_by_value[mean_sale_price_by_value == nearest_mean].index[0]
        
        # Update the missing value in the current column
        combined_data.at[index, column] = nearest_value

if combined_data.isnull().sum().sum() == 0:
    print("All missing values have been successfully updated.")
else:
    print("Some missing values remain in the dataset.")
```

## 3.10 Correlation between the four bathroom features and sales price
Given the number of variables that show the quantity of bathrooms, we can combine it into a variable with the total number of bathrooms as:

$$ \text{TotalBathrooms}=(\text{BsmtFullBath} + \text{FullBath}) + \dfrac{1}{2}(\text{BsmtHalfBath} + \text{HalfBath})$$

```python
def add_total_bathrooms(data):
    #purpose: add total bathrooms feature in dataset
    #inputted dataset needs to have 'BsmtFullBath', 'FullBath', 'BsmtHalfBath', 'HalfBath'

    with_bathrooms = data.copy()
    bath_vars = ['BsmtFullBath', 'FullBath', 'BsmtHalfBath', 'HalfBath']
    weights = pd.Series([1, 1, 0.5, 0.5], index = bath_vars)
    with_bathrooms['TotalBathrooms']=with_bathrooms[bath_vars].fillna(0)@weights
    return with_bathrooms

combined_data = add_total_bathrooms(combined_data)
```

## 3.11 Relationship between (basement size + ground living area) and sale price
It would better capture the total amount of useable area within the house, which would intuitively make the house more valuable. There might be a strong correlation between the basement square feet and the living area above ground to the sales price, so combining the variables might increase the correlation.

```python
def add_total_SF(data):
    #purpose: add total square foot variable based on total basement SF and above ground SF to dataset

    total_SF_df = data.copy()
    total_SF_df['TotalSF'] = total_SF_df['TotalBsmtSF'] + total_SF_df['GrLivArea']
    return total_SF_df

combined_data = add_total_SF(combined_data)

#scatter plot of total SF vs sales price
plt.figure(figsize=(7,3))
sns.histplot(combined_data, x='TotalSF')
plt.xlabel('Total Square Feet')
plt.ylabel("Frequency")
```

# 4. Split Data

```python
x = numeric_data.drop("SalePrice", axis=1)
y = numeric_data["SalePrice"]

print(f"Rowcount: {len(x)}")

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

print(f"Training data shape: {x_train.shape}")
print(f"Test data shape: {x_test.shape}")
```

# 5. Random Forest Regressor Model (RFRM)

## 5a. Create RFRM

```python
rf = RandomForestRegressor(n_estimators=300, criterion='friedman_mse', max_depth=30, random_state=42)
```

## 5b. Train RFRM

```python
rf.fit(x_train, y_train)
```

## 5c. Evaluate RFRM

```python
# Instantiate the RandomForestRegressor
rf = RandomForestRegressor()

# Define the number of trees to evaluate
n_trees_list = [100, 200, 300, 400, 500, 600]

# Evaluate the mean score for each number of trees using 5-fold cross-validation
for n_trees in n_trees_list:
    rf.n_estimators = n_trees  # Set the number of trees
    scores = cross_val_score(rf, x_train, y_train, cv=5)  # Perform 5-fold cross-validation
    mean_score = np.mean(scores)  # Calculate the mean score
    print(f"n_estimators={n_trees}: Mean Accuracy = {mean_score:.4f}")
```

## 5d. RFRM Hyperparameter Tuning
Based on previous result, optimal value for n_estimators parameter is 400

```python
rf = RandomForestRegressor(n_estimators=400, criterion='friedman_mse', max_depth=30, random_state=42)
rf.fit(x_train, y_train)
```

## 5e. RFRM Results
The model performed well with high accuracy scores

```python
model_score = rf.score(x_train, y_train)
print('RFRM Accuracy Training:', round(model_score,4))

model_score = rf.score(x_test, y_test)
print('RFRM Accuracy Testing:', round(model_score,4))

predictedSalePrices = rf.predict(x_test)
mse = mean_squared_error(y_test, predictedSalePrices)
print(f"RFRM Mean Squared Error Testing: {mse:.4f}")

rmse = np.sqrt(mse)
print(f"RFRM Root Mean Squared Error Testing: {rmse:.4f}")
```

#### Versus: Sum of differences in predictions for both RFR and LR

```python
predictedSalePricesRFR = rf.predict(x_test)
predictedSalePricesLR = lr.predict(x_test)

actualSalePrices = y_test

# Create DataFrame to store all sale prices
SalesPriceData = pd.DataFrame({'RFR': predictedSalePricesRFR, 'LR': predictedSalePricesLR, 'Actual': y_test})

# Calculate differences between actual and predicted sale prices for RFR and LR
SalesPriceData['Diff_Actual_RFR'] = abs(SalesPriceData['Actual'] - SalesPriceData['RFR'])
SalesPriceData['Diff_Actual_LR'] = abs(SalesPriceData['Actual'] - SalesPriceData['LR'])

# Compare which model has smaller deviation by summing the differences
total_deviation_RFR = SalesPriceData['Diff_Actual_RFR'].sum().round()
total_deviation_LR = SalesPriceData['Diff_Actual_LR'].sum().round()

# Print the total deviations
print("Sum of difference Actual vs RFR:", total_deviation_RFR)
print("Sum of difference Actual vs LR:", total_deviation_LR)

# Compare which model has smaller total deviation
if total_deviation_RFR < total_deviation_LR:
    print("\nRandom Forest Regressor (RFR) prediction has smaller total deviation.")
elif total_deviation_RFR > total_deviation_LR:
    print("\nLinear Regression (LR) prediction has smaller total deviation.")
else:
    print("\nBoth models have the same total deviation.")
```
