# Random Forrest

Random Forest is one of **Ensemble Learning** technique which is called **Bagging (Bootstrap Aggregating)** 
    
Random Forest operates by constructing: 
a **large number of decision trees** randomly, 
where **each** decision **tree** 
is **built** with **randomly** selected subsets of **data** 
and **randomly** chosen subsets of **features**.

So the overall code structure must reflect building process of the forest 
by creating, training, and combining multiple decision trees.

    
# Steps to build and test a Random Forest Regression model:

## 1. EDA (Exploratory Data Analysis). Key steps involved:

### 1.1 Data cleaning: 
Check for **missing values, inconsistencies, and errors in the data**. Clean and pre-process the data as needed.

### 1.2 Univariate analysis: 
**Analyze each variable individually** to understand its **distribution, central tendency, and variability**. 
This can involve calculating summary statistics like **mean, median, standard deviation**, and creating visualizations like **histograms and boxplots**.

### 1.3 Bivariate analysis: 
Explore the **relationships between pairs of variables**. 
This can involve creating **scatter plots, correlation matrices**, and other visualizations.

### 1.4 Multivariate analysis: 
Analyze the **relationships between multiple variables simultaneously**. 
This can involve using techniques like **principal component analysis (PCA) and factor analysis**.

## 2 Bootstrapping: 
The **train_test_split** function indirectly performs bootstrapping by taking multiple random samples (training sets) from the original data.

## 3 Decision tree training: 
The **RandomForestClassifier** object handles this internally, creating multiple decision trees and training them on different bootstrap samples.

### 3.1 Random subset of features: 
The **RandomForestClassifier parameter max_features** controls the number of features considered at each split, simulating random selection.

### 3.2 Best split among chosen features: 
This is handled automatically by the decision tree learning algorithm within the **RandomForestClassifier**.

## 4 Prediction

### 4.1 Classification: 
The **predict method** of the trained forest uses the **majority vote rule (or plurality voting) among individual trees** to predict the class label for new data points.

### 4.2 Regression: 
The **predict method** calculates the **average prediction from all trees for new data points**.

```python
def remove_excess_missing(data, upper_limit = 25):
    #purpose: remove features with too many missing values

    nan_mean = data.isna().mean()*100
    nan = nan_mean[nan_mean > upper_limit].sort_values(ascending = False)

    return data.drop(columns = nan.index)

training_data = remove_excess_missing(combined_data,25)
```

## 1.3 EDA - Bivariate Analysis

### 1.3.1 EDA - Bivariate Analysis: check correlation between the four bathroom features and sales price
Given the number of variables that show the quantity of bathrooms, we can combine it into a variable with the total number of bathrooms as:

$$ \text{TotalBathrooms}=(\text{BsmtFullBath} + \text{FullBath}) + \dfrac{1}{2}(\text{BsmtHalfBath} + \text{HalfBath})$$

```python
def add_total_bathrooms(data):
    #purpose: add total bathrooms feature in dataset
    #inputted dataset needs to have 'BsmtFullBath', 'FullBath', 'BsmtHalfBath', 'HalfBath'

    with_bathrooms = data.copy()
    bath_vars = ['BsmtFullBath', 'FullBath', 'BsmtHalfBath', 'HalfBath']
    weights = pd.Series([1, 1, 0.5, 0.5], index = bath_vars)
    with_bathrooms['TotalBathrooms']=with_bathrooms[bath_vars].fillna(0)@weights
    return with_bathrooms

combined_data = add_total_bathrooms(combined_data)
```

# 2. Bootstrapping

## 2.1 Preparing for Bootstrapping

### Count unique values for each non-numeric features (categorical features)

```python
non_numeric_features = combined_data.select_dtypes(exclude=['number']).columns.tolist()

# Create an empty DataFrame to store the count of unique values for each feature
unique_values_counts = pd.DataFrame(columns=['Feature', 'Unique Values Count'])

for feature in non_numeric_features:
    unique_values_count = combined_data[feature].nunique()
    new_row = pd.DataFrame({'Feature': [feature], 'Unique Values Count': [unique_values_count]})
    unique_values_counts = pd.concat([unique_values_counts, new_row])

print(unique_values_counts.reset_index(drop=True))
```

### Create one-hot encoded columns from each non-numeric feature

```python
# Iterate through each non-numeric feature
for feature in non_numeric_features:
    
    # Create one-hot encoded columns using get_dummies with a prefix for the current feature
    one_hot_encoded = pd.get_dummies(combined_data[feature], prefix=feature)
    
    # Join the one-hot encoded columns with the original DataFrame
    combined_data = combined_data.join(one_hot_encoded)
```

```python
combined_data.columns.values
```

### New dataframe "numeric_data" that encode numeric features from original dataframe

```python
# Get a list of all column names in the DataFrame
all_column_names = combined_data.columns.values.tolist()

# Create a list of column names to keep (numeric features)
numeric_features_to_keep = [column_name for column_name in all_column_names if column_name not in non_numeric_features]

# Retain only the columns with numeric data types in the DataFrame
numeric_data = combined_data[numeric_features_to_keep]
```

# 3. Decision Tree Training

```python
rf = RandomForestRegressor(n_estimators=150, criterion='friedman_mse', max_depth=30, random_state=42)
rf.fit(x_train, y_train)
```

```python
# Assuming rf is your RandomForestClassifier
scores = cross_val_score(rf, x_train, y_train, cv=5)  # 5-fold cross-validation

# Evaluate the mean score for each number of trees
for n_trees, score in zip([50, 100, 150, 200, 1000], scores):
    print(f"n_estimators={n_trees}: Mean Accuracy = {score:.4f}")
```

# 4. Evaluate Model

```python
model_score = rf.score(x_train, y_train)
print('Model Accuracy Training:', round(model_score,4))

model_score = rf.score(x_test, y_test)
print('Model Accuracy Testing:', round(model_score,4))
```
